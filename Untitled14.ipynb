```python
import pandas as pd
```

```python
data = pd.read_csv("dummy_sample.csv")
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
top_fund = data.groupby('Fund')['Investment ($M)'].sum().nlargest(5)
top_fund
```

```python
top_com = data.groupby('Company Name')['Total Emissions by Fund (tons of CO2e)'].sum().nlargest(5)
top_com
```

```python
top_country = data.groupby('Country')['Country Capital Catalyzed ($M)'].sum().nlargest(5)
top_country
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x=top_fund.index, y=top_fund.values)
plt.title('Top 5 Funds by Investment')
plt.ylabel('Total Investment ($M)')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x=top_com.index, y=top_com.values)
plt.title('Top 5 Companies by Emissions')
plt.ylabel('Total Emissions (tons of CO2e)')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x=top_country.index, y=top_country.values)
plt.title('Top 5 Countries by Capital Catalyzed')
plt.ylabel('Total Capital Catalyzed ($M)')
plt.show()
```

```python
numerical_data = data.select_dtypes(include=['float64', 'int64'])
```

```python
correlation = numerical_data.corr()
plt.figure(figsize=(10,8))
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title('Correlation between Numerical Features')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=data, x='Fund Size ($M)', y='Investment ($M)', hue='Fund')
plt.title('Fund Size vs Investment')
plt.xlabel('Fund Size ($M)')
plt.ylabel('Investment ($M)')
plt.show()
```
